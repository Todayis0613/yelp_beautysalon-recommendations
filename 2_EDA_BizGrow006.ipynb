### Summary Statistics of the Data

```python
import pandas as pd
import spacy

# Load spaCy's English NLP model
nlp = spacy.load("en_core_web_sm")

# Function to tokenize text using spaCy
def tokenize_text(text):
    doc = nlp(str(text).lower())  # Convert text to lowercase and tokenize
    return [token.text for token in doc if not token.is_punct and not token.is_space]

# Apply tokenization to reviews
reviews_df['tokens'] = reviews_df['text'].apply(tokenize_text)

# **A. Number of reviews**
num_reviews = len(reviews_df)

# **B. Total number of tokens**
total_tokens = sum(reviews_df['tokens'].apply(len))

# **C. Number of unique words (vocabulary size)**
unique_tokens = set()
reviews_df['tokens'].apply(lambda tokens: unique_tokens.update(tokens))
vocabulary_size = len(unique_tokens)

# **D. Average review length (average number of words per review)**
average_review_length = total_tokens / num_reviews

# **E. Number of unique customers**
num_unique_customers = reviews_df['user_id'].nunique()

# **F. Other key elements**
num_businesses = reviews_df['business_id'].nunique()  # Number of businesses covered
avg_stars = reviews_df['stars'].mean()  # Average star rating per review
avg_votes_per_review = reviews_df[['useful', 'funny', 'cool']].sum(axis=1).mean()  # Average total votes per review

# Print results
print(f"Number of reviews: {num_reviews}")
print(f"Total number of tokens: {total_tokens}")
print(f"Number of unique words (vocabulary size): {vocabulary_size}")
print(f"Average review length: {average_review_length:.2f} words per review")
print(f"Number of unique customers: {num_unique_customers}")
print(f"Number of businesses covered: {num_businesses}")
print(f"Average stars per review: {avg_stars:.2f}")
print(f"Average votes per review: {avg_votes_per_review:.2f}")
```

```python
# Function to process text using spaCy: Tokenization & Lemmatization
def process_text_spacy(text):
    doc = nlp(text)  # Process text using spaCy NLP pipeline
    tokens = [token.text for token in doc]  # Tokenization
    lemmatized = [token.lemma_ for token in doc]  # Lemmatization
    return tokens, lemmatized
```

```python
# Select a sample review
sample_review = one_star_reviews[0]  # Take the first 1-star review

# Process text
tokens, lemmatized = process_text_spacy(sample_review)

# Print results
print("Original Text:", sample_review)
print("Tokenized:", tokens)
print("Lemmatized:", lemmatized)
```

### Top 20 frequently used nouns in 1 star reviews and 5 star reviews

```python
from collections import Counter

# Function to extract top nouns
def get_top_nouns(texts, top_n=20):
    word_counts = Counter()  # Initialize counter

    for text in texts:
        doc = nlp(text)  # Process each review
        words = [token.lemma_ for token in doc if token.pos_ == "NOUN"]  # Extract nouns
        word_counts.update(words)  # Count occurrences

    return word_counts.most_common(top_n)  # Return top N nouns

# Get top nouns for 1-star and 5-star reviews
top_nouns_1_star = get_top_nouns(one_star_reviews)
top_nouns_5_star = get_top_nouns(five_star_reviews)

# Print results
print("\nTop 20 Nouns in 1-star reviews:", top_nouns_1_star)
print("Top 20 Nouns in 5-star reviews:", top_nouns_5_star)
```

## VADER Sentiment Analysis

```python
!pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
```

```python
# Initialize Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment scores
def get_sentiment(text):
    sentiment_score = analyzer.polarity_scores(text)
    return sentiment_score["compound"]

# Apply sentiment analysis to 1-star and 5-star reviews
reviews_df["sentiment_score"] = reviews_df["text"].apply(get_sentiment)

# Classify sentiment into categories
def classify_sentiment(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

reviews_df["sentiment_category"] = reviews_df["sentiment_score"].apply(classify_sentiment)
```

```python
# Show sentiment distribution
print(reviews_df["sentiment_category"].value_counts())
```

```python
# Define an orange gradient color palette
orange_palette = sns.color_palette("Oranges", 3)  # Generates 3 shades of orange

plt.figure(figsize=(8, 5))
sns.countplot(data=reviews_df, x="sentiment_category",
              order=["Negative", "Neutral", "Positive"],
              palette=orange_palette)

plt.xlabel("Sentiment Category", fontsize=12)
plt.ylabel("Number of Reviews", fontsize=12)
plt.title("Sentiment Distribution of Reviews", fontsize=14)
plt.show()
```

```python
# Set style to match the polished look
sns.set_style("darkgrid")

# Set figure size
plt.figure(figsize=(12, 7))

# Create histogram with KDE curve
sns.histplot(reviews_df["sentiment_score"], bins=50, kde=True,
             color="steelblue", edgecolor="black", alpha=0.6,
             line_kws={"color": "navy", "linewidth": 2})

# Labels and title formatting
plt.xlabel("Sentiment Score", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.title("Distribution of Sentiment Scores", fontsize=16)

# Show the plot
plt.show()
```
